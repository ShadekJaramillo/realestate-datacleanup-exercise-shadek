#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
# a that I used during the some of the next exercises

def compare(ds_1:pd.DataFrame, ds_2:pd.DataFrame, column_to_compare:str) -> dict:
    ret = {} #dictionary that will keep the variable for the return.

    average_price_ds_1 = ds_1[column_to_compare].mean()
    average_price_ds_2 = ds_2[column_to_compare].mean()

    if average_price_ds_1 >= average_price_ds_2:
        higher_price_region, higher_price_average = ds_1, average_price_ds_1
        lower_price_region, lower_price_average =  ds_2, average_price_ds_2
    else:
        higher_price_region, higher_price_average = ds_2, average_price_ds_2
        lower_price_region, lower_price_average = ds_1, average_price_ds_1
    
    price_difference = higher_price_average - lower_price_average
    percentage_difference = (price_difference/lower_price_average)*100

    return {
        'average price ds_1': average_price_ds_1,
        'average price ds_2': average_price_ds_2,
        'higher price region': higher_price_region,
        'higher price average': higher_price_average,
        'lower price region': lower_price_region,
        'lower price average': lower_price_average,
        'percentage difference': percentage_difference,
        'price difference': price_difference
    }
```

```python
# I used the commented lines of code below to find if there are other instances of Valdemorillo or Galapagar in other levels.
# All instances of valdemorillo and Galapagar are in the level5 column.

for i in range(1,9):
    level = ds['level'+str(i)]
    print(level[level == 'Galapagar' ])
    print(level[level == 'Valdemorillo' ])

```

```python
# TODO

ds_Valdemorillo = ds[ds.level5 == 'Valdemorillo']
ds_Galapagar = ds[ds.level5 == 'Galapagar']


aux = compare(ds_Valdemorillo, ds_Galapagar, 'price')

average_price_Valdemorillo = aux['average price ds_1']
average_price_Galapagar = aux['average price ds_2']
higher_price_region = aux['higher price region']
higher_price_average = aux['higher price average']
lower_price_region = aux['lower price region']
lower_price_average = aux['lower price average']
percentage_difference = aux['percentage difference']
standard_deviation_Valdemorillo = ds_Valdemorillo.price.std()
standard_deviation_Galapagar = ds_Galapagar.price.std()


print(f'Average price in Valdemorillo: {average_price_Valdemorillo:,.2f} EUR')
print(f'Average price in Galapagar: {average_price_Galapagar:,.2f} EUR')
print(f'The average difference in price is {abs(average_price_Galapagar - average_price_Valdemorillo):,.2f} EUR')
print(f'The price in {higher_price_region.iloc[0].level5} is {percentage_difference:,.2f}% higher than in {lower_price_region.iloc[0].level5}')
print(f'The standard deviation of the price in Valdemorillo is of {standard_deviation_Valdemorillo:,.2f} EUR')
print(f'The standard deviation of the price in Galapagar is of {standard_deviation_Galapagar:,.2f} EUR')
```

##### **conclusion**

Although the price difference is $\approx$ 3800 EUR and that's not a small amount of money, it represents only around a 1% difference in the average price between the regions of Valdemorillo and Galapagar. Given that the standar deviation in both regions is quite high, the relevance of that difference is even smaller, since when picking a random house on either Valdemorillo or Galapagar, odd are its price will be far from both average values by way more than the price difference between the regions.
